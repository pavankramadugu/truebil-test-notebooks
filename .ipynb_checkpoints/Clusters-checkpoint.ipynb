```python
status_query = ("""select * from 
(
  select
          cj.listing_id,
          cj.dt,
          case when cj.dt between start_time and end_time then act.status end as status
        from
        (
          select
            distinct ls.id as listing_id,
            dt.dt as dt
          from
            (
              select
                id
              from
                listings
              where
              is_inventory = 1
              and created_at >= date_add(current_date,interval -12 month) 
            ) ls
          cross join 
            (
              select
                distinct date(created_at) as dt
              from
                buyer_listings
              where
                created_at >= date_add(current_date,interval -12 month) 
            ) dt 
        )cj
        left join 
        (
          select
            listing_id,
            status,
            start_time,
            case when end_time is null then date('2100-01-01') else end_time end as end_time
      from
      (
        SELECT
l.listing_id,
l.status_to AS status,
      l.changed_at AS start_time,
      CASE
        WHEN r.changed_at IS NULL THEN NOW()
        ELSE r.changed_at
      END AS end_time
    FROM
      (
        SELECT @row_no1 := @row_no1 + 1 AS row_no,
        listing_id,
        status_to,
        changed_at
      FROM
        listing_status_track,
        (
          SELECT @row_no1 := 0) t
      ORDER BY
        listing_id ,
        changed_at) l
    LEFT JOIN (
        SELECT @row_no2 := @row_no2 + 1 AS row_no,
        listing_id,
        status_to,
        changed_at
      FROM
        listing_status_track,
        (
          SELECT @row_no2 := 0) p
      ORDER BY
        listing_id ,
        changed_at) r ON
      l.row_no + 1 = r.row_no
      AND l.listing_id = r.listing_id
      ) lf
      where
        listing_id is not null
        ) act on cj.listing_id = act.listing_id 
        left join listings ls on cj.listing_id = ls.id
        where ls.is_inventory = 1
        and ls.inventory_type_id = 2
  )a
  where 0=0
  and status is not null""")
listing_status = sql_result(status_query)
```

```python
sold_query = ("""select t6.id, t6.lpc, t6.token, DATE_ADD(t6.lpc, interval 3 day) AS lpcp3, DATE_ADD(date(t6.token), 
                interval -1 day) AS tokenm1, lo.city_id, l.body_type_id, l.price, l.fuel_type_id_primary, 
                l.transmission_type_id, (YEAR(t6.token) - YEAR(l.registration_date)) AS age, datediff(date(t6.token), 
                t6.lpc) as dif, m.id as model 
                from
                (select t5.id, coalesce(t5.lpc, t5.active_date) as lpc, t5.token from
                    (select t3.id, t3.lpc, t3.token, date(t4.Date) as active_date from
                        (select t2.id, date(max(lpc.created_at)) as lpc, t2.token from
                            (select t1.id, t1.token from
                                (select l.id, max(date(lsl.start_time)) as token from listings l join 
                                listing_status_lifeline lsl on l.id = lsl.listing_id where lsl.status = 'truebiled' 
                                group by l.id) t1 where t1.token >= date_add(current_date,interval -12 month)) t2
                            left join listing_price_changes lpc on t2.id = lpc.listing_id 
                            where coalesce(lpc.created_at,'2010-01-01') <= t2.token group by t2.id) t3 
                        left join 
                            (select 'Carid','Status','Date' union select listing_id, lifeline.`status`, 
                            if( min(start_time)< inventory_added_date, inventory_added_date, min(start_time)) as 
                            first_active_date from (select l.listing_id, l.status_to as `status`, l.changed_at as 
                            start_time, case when r.changed_at is null then now() else r.changed_at end as end_time from 
                            (select @row_no1:=@row_no1 + 1 as row_no, listing_id, status_to, changed_at 
                            from listing_status_track, (select @row_no1:=0) t order by listing_id, changed_at) l 
                            left join (select @row_no2:=@row_no2 + 1 as row_no, listing_id, status_to, changed_at 
                            from listing_status_track, (select @row_no2:=0) p order by listing_id, changed_at) r 
                            on l.row_no + 1=r.row_no and l.listing_id=r.listing_id) lifeline left join listings 
                            on listings.id=lifeline.listing_id and lifeline.end_time>=listings.inventory_added_date 
                            where lifeline.`status` in ('Active','Refurbishing') group by listing_id) t4
                        on t3.id = t4.Carid) t5)
                    t6
                join listings l on l.id = t6.id join localities lo on lo.id = l.locality_id join variants v 
                on l.variant_id = v.id join models m on m.id = v.model_id where t6.token >= t6.lpc;""")
sold_cars1 = sql_result(sold_query)
```

```python
crf_query = ("""select t1.id as id, date(t1.created_at) as date, count(t1.buyer_id) as CRF from
                (SELECT b.id as buyer_id, l.id, bl.created_at as created_at FROM listings l join buyer_listings bl 
                on bl.listing_id = l.id join buyers b on b.id = bl.buyer_id where l.is_inventory = 1) t1
                group by t1.id, date(t1.created_at);""")
crf_data = sql_result(crf_query)
```

```python
mo_query = ("""select t1.id as id, date(t1.created_at) as date, count(t1.buyer_id) as MO from
                (SELECT b.id as buyer_id, l.id, blo.created_at as created_at FROM listings l join buyer_listing_offers blo 
                on blo.listing_id = l.id join buyers b on b.id = blo.buyer_id where l.is_inventory = 1) t1
                group by t1.id, date(t1.created_at);""")
mo_data = sql_result(mo_query)
```

```python
crf_weight = 4
mo_weight = 10.3
s = crf_weight + mo_weight
# crf_weight = 1
# mo_weight = 1
crf_weight = 2*crf_weight/s
mo_weight = 2*mo_weight/s
```

```python
for i in sold_cars.id.unique():
    
    listing_data = listing_status[listing_status['listing_id']==i]
    dates = sold_cars[sold_cars['id']==i]
    
    live_data = listing_data[(listing_data['status']=='active') | (listing_data['status']=='refurbishing')]
    
    live_data['flag'] = np.where((live_data['dt']>dates.lpcp3.iloc[0])&(live_data['dt']<=dates.tokenm1.iloc[0]),1,None)
    
    live_data = live_data[live_data['flag']==1]
    live_data = live_data.head(7)
    
#     holdup = len(live_data[(live_data['dt']>dates.lpcp3.iloc[0]) & (live_data['dt']<=dates.tokenm1.iloc[0])])
    holdup = len(live_data)
    
    city = dates.city_id.iloc[0]
    
    crf_raw = crf_data[crf_data['id']==i]
    crf = (crf_raw[(crf_raw['date']>dates.lpcp3.iloc[0]) & (crf_raw['date']<=dates.tokenm1.iloc[0])]).CRF.sum()
    
    mo_raw = mo_data[crf_data['id']==i]
    mo = (mo_raw[(mo_raw['date']>dates.lpcp3.iloc[0]) & (mo_raw['date']<=dates.tokenm1.iloc[0])]).MO.sum()
    
#     lps = ((crf*crf_weight + mo*mo_weight)/(crf_weight + mo_weight))/holdup
    lps = (crf*crf_weight + mo*mo_weight)/holdup
    
    bd = dates.body_type_id.iloc[0]
    price = dates.price.iloc[0]
    fuel = dates.fuel_type_id_primary.iloc[0]
    tr = dates.transmission_type_id.iloc[0]
    age = dates.age.iloc[0]
    mod = dates.model.iloc[0]
    
    final = final.append({'id': i, 'city': city, 'holdup': holdup, 'crf': crf, 'mo': mo, 'lps': lps,
                          'bd': bd, 'price': price, 'fuel': fuel, 'tr': tr, 'age': age, 'model':mod} , ignore_index=True)
```

```python
def fifth(df, price_level, bd_level, fuel_level, age_level):
    global min_pts
    global percentile
    
    age_level = str(int(age_level))
    if(len(df)>=min_pts):
        if(len(df[df['t']==1])>=min_pts):
            
            df1 = df[df['t']==1]
            df2 = df1[df1['lps']<np.percentile(df1.lps,percentile)]
            
            vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t1" + "a"] = df2.lps.mean()
            vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t1" + "d"] = df2.lps.std()
            
        else:
            vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t1" + "a"] = None
            vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t1" + "d"] = None
            
        print("p",price_level,"b",bd_level,"f",fuel_level,"a",age_level,"t","1 = ",vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t1" + "a"], vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t1" + "d"])
            
        if(len(df[df['t']==2])>=min_pts):
            
            df1 = df[df['t']==2]
            df2 = df1[df1['lps']<np.percentile(df1.lps,percentile)]
            
            vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t2" + "a"] = df2.lps.mean()
            vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t2" + "d"] = df2.lps.std()
            
        else:
            vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t2" + "a"] = None
            vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t2" + "d"] = None
        
        print("p",price_level,"b",bd_level,"f",fuel_level,"a",age_level,"t","2 = ",vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t2" + "a"], vars()["p" + price_level + "_" + "b" + bd_level + "f" + fuel_level + "_" + "a" + age_level + "_" + "t2" + "d"])
    else:
        print("Not enough points of price level ", price_level, " of bodytype level ", bd_level, " of fuel level ", fuel_level, " of age level ", age_level, " at transmission level")
        return()
```
